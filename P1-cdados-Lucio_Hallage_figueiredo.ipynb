
# Hipótese
*Um país que entra no mercosul aumenta o número de Exportações em % do PIB(GDP).*
# Mecanismo
*Já que o Mercosul é um mercado comum entre seus países membros, ele garante a livre circulação de exportações em seus paises membros, sendo assim aumentando suas exportações.*

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
#matplotlib.style.use("classic")
```

```python
dados = pd.read_excel('Exports.xlsx')
data = pd.read_excel('Hightech_export.xlsx')
```

```python
dados.head()
```

```python
dadost=dados.set_index('Exports of goods and services (% of GDP)')
data=data.set_index('High-technology exports (% of manufactured exports)')
dadost.columns
```

```python
dadost.loc[:,['2009','2010']]
```

```python
mercosul=['Brazil','Argentina','Uruguay','Paraguay']
i=1988
anos=[]
while i<2012:
    a=str(i)
    anos.append(a)
    i+=1
    
```

```python
dadost.loc[mercosul,anos]
```

```python
for i in mercosul:
    plt.plot(dadost.loc[i,anos],label=i)
plt.axis(['1988','1996',0,60])
plt.xlabel('Anos')
plt.ylabel('Exportações em % do PIB(GDP)')
plt.legend()
plt.show()
```

O mercosul foi criado em 1991 e parece que os países não tiveram mudanças significativas nas suas exportações.Agora vamos ver os países que entraram depois da sua criação.

```python
associados=['Chile','Peru','Colombia','Ecuador']
```

Chile entra em 1996.
Peru entra em 2003.
Colombia e Equador entram em 2004.

```python
for i in associados:
    
    plt.plot(dadost.loc[i,anos],label=i)
    plt.axis(['2001','2011',10,40])
    plt.xlabel('Anos')
    plt.ylabel('Exportações em % do PIB(GDP)')
    plt.legend()
    if i=="Chile":
        plt.axis(['1994','2007',25,45])
        plt.show()
plt.show()
```

Os paises membros do Mercosul não tiveram um crescimento significativo quando criaram o mercosul, porém Os países associados quando entraram no mercosul tiveram um aumento significativo em suas exportações, esse fato deve acontecer pelo Mercosul ser um mercado comum entre países.

```python
s=0
anost=[]
expm=[]
for i in dadost.columns:
    if int(i)>1987:
        #print('No ano {1} a media de exportações em % do PIB(GDP) é {0}'.format(dadost.loc[:,i].mean(),i))
        anost.append(i)
        expm.append(dadost.loc[:,i].mean())
```

```python
for i in associados:
    plt.plot(anost,expm,label='mundial')
    plt.plot(dadost.loc[i,anos],label=i)
    if i=="Chile":
        plt.axis(['1994','2007',0,47.5])
    if i=="Peru":
        plt.axis(['2001','2011',0,47.5])
    if i=="Colombia":
        plt.axis(['2001','2011',0,47.5])
    if i=="Ecuador":
        plt.axis(['2001','2011',0,47.5])
    print(i)
    plt.xlabel('Anos')
    plt.ylabel('Exportações em % do PIB(GDP)')
    plt.legend()
    plt.show()
```

Agora comparando com a exportação media dos outros países podemos ver que a media mundial acabou tendo um bom aumento.

Vamos comparar as taxas de crescimento anuais para melhorar nossa visualização do problema.

```python
def tx_de_crescimento(pais):
    cresexp=[0]
    i=0
    while i<(len(pais)-1):
        cresexp.append(100*(pais[i+1]-pais[i])/pais[i])
        i+=1
    return cresexp

mundial_cres=tx_de_crescimento(dadost.loc[:,anos].mean())
chile_cres=tx_de_crescimento(dadost.loc['Chile',anos])
peru_cres=tx_de_crescimento(dadost.loc['Peru',anos])
equador_cres=tx_de_crescimento(dadost.loc['Ecuador',anos])
colombia_cres=tx_de_crescimento(dadost.loc['Colombia',anos])
associados_cres=tx_de_crescimento(dadost.loc['Colombia',anos])

plt.plot(anos,mundial_cres,label='Mundo')
plt.plot(anos,chile_cres,label='Chile')
plt.xlabel('Anos')
plt.ylabel('Taxa de crescimento das Exportações em % do PIB(GDP)')
plt.legend()
plt.axis(['1992','2004',-30,30])
plt.show()

plt.plot(anos,mundial_cres,label='Mundo')
plt.plot(anos,peru_cres,label='Peru')
plt.plot(anos,equador_cres,label='Ecuador')
plt.plot(anos,colombia_cres,label='Colombia')
plt.xlabel('Anos')
plt.ylabel('Taxa de crescimento das Exportações em % do PIB(GDP)')
plt.legend()
plt.axis(['2000','2010',-30,30])
plt.show()
```

```python
d = {'Associados':associados_cres, 'Mundo': mundial_cres}
df = pd.DataFrame(data=d)
df.boxplot()
```

Podemos ver que apos a data de 1996(ano que o Chile entrou no mercosul) as taxas de crescimento do Chile ficaram boas e um pouco melhores que a do crescimento medio mundial.

A taxa de crescimento do Peru a partir de 2003(ano que o Peru entrou no mercosul) ficaram muito superiores a media mundial como a do Ecuador a partir de 2004(ano que o Ecuador e Colombia entraram no mercosul),só a Colombia q parece acompanhar a media mundial.

Em 2009 aconteceu uma crise que ocasionou as quedas nas taxas de crescimento de todos os países e do mundo.
